```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\mgoyal\Documents\2007\past_prices.csv\past_prices.csv")
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df['orig'] = le.fit_transform(df['orig']) 
df['dest'] = le.fit_transform(df['dest'])
df['dep_date'] = le.fit_transform(df['dep_date']) 
df['ret_date'] = le.fit_transform(df['ret_date'])
df['airline'] = le.fit_transform(df['airline']) 
df['search_date'] = le.fit_transform(df['search_date']) 
```

```python
df
```

```python
y = df['price']
y
```

```python
df2 = df
df2
```

```python
df.drop(['price'], axis = 1, inplace = True)
```

```python
df
```

```python
X2 = np.array(df)
y2 = np.array(y)
```

```python
X2
```

```python
y2
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X2, y2, test_size = 0.25, random_state = 42)
```

```python
# Import the Random Forest model
from sklearn.ensemble import RandomForestRegressor

# Instantiate model with 20 decision trees
rf = RandomForestRegressor(n_estimators = 20,random_state = 42)

# Train the model on training data
rf.fit(X_train, y_train);
```

```python
preds_rf = rf.predict(X_test)
```

```python
errors = abs(preds_rf - y_test)
```

```python
mape = 100*(errors/ y_test)
accuracy = 100 - np.mean(mape)
print('Accuracy:)', round(accuracy,2), '%.')
```
